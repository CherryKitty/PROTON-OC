```python
import pandas as pd
from ast import literal_eval
```

```python
def format_list(x):
    return literal_eval(x.replace(" ", ","))

data = pd.read_csv("table-output.csv", 
                   skiprows=5, 
                   header=1, 
                   names=['run', 'persons', 'oc-persons', 'scenario', 'targets-addressed-percent',\
                          'ticks-between-intervention', 'intervention-start', 'intervention-end',\
                          'OC-members-scrutinize?', 'OC-arrest-multiplier', 'OC-members-repression?',\
                          'OC-members-per-tick', 'tick', 'OC members', 'recruited individuals into OC',\
                          'crimes', 'crimes by OC members', 'C', 'R', 'new recruited edu lvl',\
                          'new recruited wealth lvl', 'new recruited job lvl', 'OC members edu lvl',\
                          'OC members wealth lvl', 'OC members job lvl', 'ordinary criminals edu lvl',\
                          'ordinary criminals wealth lvl', 'ordinary criminals job lvl',\
                          'population edu lvl', 'population wealth lvl', 'population job lvl',\
                          'crimes committed by facilitators', 'law enforcement interventions'],
                   converters={
                       "C": lambda x: format_list(x),
                       "R": lambda x: format_list(x),
                       'new recruited edu lvl': lambda x: format_list(x),
                       'new recruited wealth lvl': lambda x: format_list(x), 
                       'new recruited job lvl': lambda x: format_list(x),
                       'OC members edu lvl': lambda x: format_list(x), 
                       'OC members wealth lvl': lambda x: format_list(x),
                       'OC members job lvl': lambda x: format_list(x), 
                       'ordinary criminals edu lvl': lambda x: format_list(x),
                       'ordinary criminals wealth lvl': lambda x: format_list(x),
                       'ordinary criminals job lvl': lambda x: format_list(x),
                       'population edu lvl': lambda x: format_list(x),
                       'population wealth lvl': lambda x: format_list(x),
                       'population job lvl': lambda x: format_list(x)})
data.head(1)
```

```python
# months*repetitions*population
months=len(pd.unique(data['tick']))
population=len(pd.unique(data['persons']))
repetitions=len(pd.unique(data['run']))/population # add more variables in case
```

```python
print("If the difference between data length and parameters combination is 0,\nthen \
we have all the possible combinations in the data, otherwise something went wrong.\n\
Difference is: {}".format(len(data) - months*repetitions*population))
```
